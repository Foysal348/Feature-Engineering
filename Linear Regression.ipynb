<h4>1.Simple Linear Regression</h4>

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression

#x=np.array([1,2,3,4,5,8]).reshape(-1, 1) or
x=np.array([[6],[8],[12],[14],[18]])
y=np.array([350,775,1150,1395,1675])

model=LinearRegression()
model.fit(x,y)

y_pred=model.predict(x)
y_pred

plt.scatter(x,y,color='r',label='Actual Data')
plt.plot(x,y_pred,color='g',label='Line')
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid()
plt.show()

 #display Coefficients
print("Coefficient(slope):",model.coef_)
print("Intercept:", model.intercept_)
```
