*1.Normalization*

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt 

#Initialize Data
np.random.seed(42)
feature_1=np.array([[20,31],[13,34],[13,17],[25,27]])#created a 2D array (4 samples,that have 2 features)

print(np.shape(feature_1))
##Initialize Normalization
Scaler=MinMaxScaler()
Normalized_data=Scaler.fit_transform(feature_1)

print(Normalized_data)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(feature_1[:,0],feature_1[:,1],color='green',edgecolors='r')
plt.title('Original Data')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(1,2,2)
plt.scatter(Normalized_data[:, 0], Normalized_data[:, 1], color='blue',edgecolors='y')
plt.title("Normalized Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#Sample of data 
np.random.seed(42)
feature_1=np.random.uniform(10,50,100)
feature_2=np.random.uniform(10,40,100)
data=np.column_stack((feature_1,feature_2))

#Normalized Data

Normalized_data = MinMaxScaler().fit_transform(data)
#print(Normalized_data)

#Plot Before and after Normalization
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,6))

#Before Normalization
ax1.scatter(data[:,0],data[:,1],edgecolors='blue',color='y')
ax1.set_title("Before Normalization")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")

#After Normalization
ax2.scatter(Normalized_data[:,0],Normalized_data[:,1],color='g',edgecolors='black')
ax2.set_title("After Normalization")
ax2.set_xlabel("Feature 1")
ax2.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create random data
np.random.seed(0)
data = pd.DataFrame({
    'Feature_A': np.random.randint(1, 100, 50),
    'Feature_B': np.random.normal(50, 20, 50)
})

# Min-Max Scaling
data['Feature_A_Scaled'] = (data['Feature_A'] - data['Feature_A'].min()) / (data['Feature_A'].max() - data['Feature_A'].min())
data['Feature_B_Scaled'] = (data['Feature_B'] - data['Feature_B'].min()) / (data['Feature_B'].max() - data['Feature_B'].min())

# Scatter plot before and after scaling
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(data=data, x='Feature_A', y='Feature_B', ax=axes[0])
axes[0].set_title("Original Data")

sns.scatterplot(data=data, x='Feature_A_Scaled', y='Feature_B_Scaled', ax=axes[1])
axes[1].set_title("Min-Max Scaled Data")

plt.tight_layout()
plt.show()

# Histogram comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(data['Feature_A'], kde=True, ax=axes[0], label='A')
sns.histplot(data['Feature_B'], kde=True, ax=axes[0], label='B', color='orange')
axes[0].set_title("Original Distributions")
axes[0].legend()

sns.histplot(data['Feature_A_Scaled'], kde=True, ax=axes[1], label='A (scaled)')
sns.histplot(data['Feature_B_Scaled'], kde=True, ax=axes[1], label='B (scaled)', color='orange')
axes[1].set_title("Scaled Distributions")
axes[1].legend()
#Plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Generate random data for two features
np.random.seed(0)
feature_A = np.random.randint(10, 90, 60)
feature_B = np.random.randint(0, 100, 60)
data = np.column_stack((feature_A, feature_B))

# Min-Max normalization
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
feature_A_scaled = data_scaled[:, 0]
feature_B_scaled = data_scaled[:, 1]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plots
axs[0, 0].scatter(feature_A, feature_B)
axs[0, 0].set_title("Original Data")
axs[0, 0].set_xlabel("Feature_A")
axs[0, 0].set_ylabel("Feature_B")

axs[0, 1].scatter(feature_A_scaled, feature_B_scaled)
axs[0, 1].set_title("Min-Max Scaled Data")
axs[0, 1].set_xlabel("Feature_A_Scaled")
axs[0, 1].set_ylabel("Feature_B_Scaled")

# Histograms with KDE
sns.histplot(feature_A, kde=True, bins=10, color='steelblue', alpha=0.6, label='A', ax=axs[1, 0])
sns.histplot(feature_B, kde=True, bins=10, color='orange', alpha=0.6, label='B', ax=axs[1, 0])
axs[1, 0].set_title("Original Distributions")
axs[1, 0].legend()

sns.histplot(feature_A_scaled, kde=True, bins=10, color='steelblue', alpha=0.6, label='A (scaled)', ax=axs[1, 1])
sns.histplot(feature_B_scaled, kde=True, bins=10, color='orange', alpha=0.6, label='B (scaled)', ax=axs[1, 1])
axs[1, 1].set_title("Scaled Distributions")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
#Import
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

#Generate Data
np.random.seed(0)
feature_A=np.random.randint(0,50,100)
feature_B=np.random.randint(0,50,100)
data=np.column_stack((feature_A,feature_B))#because ,take only one kwrs

#Normalization
model=StandardScaler()
data=model.fit_transform(data)
feature_A_scaled=data[:,0]
feature_B_scaled=data[:,1]

#Create Subplots
fig,axes=plt.subplots(2,2,figsize=(12,6))

#Scatter plot
#1st
axes[0,0].scatter(feature_A,feature_B,color='b',edgecolor='black')
axes[0,0].set_title('Before Standerdization')
axes[0,0].set_xlabel('Feature A')
axes[0,0].set_ylabel('Feature B')
#2nd
axes[0,1].scatter(feature_A_scaled,feature_B_scaled,color='r',edgecolor='g')
axes[0,1].set_title('After Standerdization')
axes[0,1].set_xlabel('Scaled_Feature A')
axes[0,1].set_ylabel('Scaled_Feature B')

#Histogram plot
#1st
sns.histplot(feature_A,label='Feature_A',alpha=0.7,color='orange',kde=True,bins=10,ax=axes[1,0])
sns.histplot(feature_B,label="Feature_B",alpha=0.7,color='red',kde=True,bins=10,ax=axes[1,0])
axes[1,0].set_title('Origin Distribution')
#2nd
sns.histplot(feature_A_scaled,label='B (scaled)',alpha=.7,kde=True,bins=10,ax=axes[1,1],color='orange')
sns.histplot(feature_B_scaled,label='B (scaled)',alpha=.7,kde=True,bins=10,ax=axes[1,1],color='red')
axes[1,1].set_title('Scaled Distribution')
plt.tight_layout()
axes[1,0].legend()
axes[1,1].legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Generate random data for two features
np.random.seed(0)
feature_A = np.random.randint(0, 50, 100)
feature_B = np.random.randint(0, 50, 100)
data = np.column_stack((feature_A, feature_B))

# Standardization (Z-score scaling)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
feature_A_scaled = data_scaled[:, 0]
feature_B_scaled = data_scaled[:, 1]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plots
axs[0, 0].scatter(feature_A, feature_B)
axs[0, 0].set_title("Original Data")
axs[0, 0].set_xlabel("Feature_A")
axs[0, 0].set_ylabel("Feature_B")

axs[0, 1].scatter(feature_A_scaled, feature_B_scaled)
axs[0, 1].set_title("Standardized Data")
axs[0, 1].set_xlabel("Feature_A_Standardized")
axs[0, 1].set_ylabel("Feature_B_Standardized")

# Histograms with KDE
sns.histplot(feature_A, kde=True, bins=10, color='steelblue', alpha=0.6, label='A', ax=axs[1, 0])
sns.histplot(feature_B, kde=True, bins=10, color='orange', alpha=0.6, label='B', ax=axs[1, 0])
axs[1, 0].set_title("Original Distributions")
axs[1, 0].legend()

sns.histplot(feature_A_scaled, kde=True, bins=10, color='steelblue', alpha=0.6, label='A (standardized)', ax=axs[1, 1])
sns.histplot(feature_B_scaled, kde=True, bins=10, color='orange', alpha=0.6, label='B (standardized)', ax=axs[1, 1])
axs[1, 1].set_title("Standardized Distributions")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
